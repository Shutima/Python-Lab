```python
import os

def check_my_files(array_files):
    for file in array_files:
        if os.path.exists(file):
            print ("File exist: {0}".format(file))
        else:
            print ("File does not exist: {0}".format(file))
```

```python
check_my_files(array_names)
```

```python
# try this as well https://www.techbeamers.com/pandas-merge-csv-files/
from glob import glob
def merge_files(arr_files, separator= ',', concat_axis = 0):
    data_f = []
    for file in arr_files:
        data_f.append(pd.read_csv(file, sep = separator,index_col=0))
    res = pd.concat(data_f, axis=concat_axis).reset_index(drop=True)
    return res
```

```python
merge_files(array_names)
```

```python
def read_array_csv_files(arr_files, separator = ',', concat_axis=0):
    data_f = []
    for file in arr_files:
        data_f.append(pd.read_csv(file, sep=separator, index_col=0))
    res = pd.concat(data_f, axis = concat_axis).reset_index(drop=True)
    return res
```

```python
df_iris_from_disk = read_array_csv_files(array_names)
```
